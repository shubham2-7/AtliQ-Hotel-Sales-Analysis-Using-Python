## ==> 3. Data Transformation

```python
fact_agg_bookings.head()
```

```python
fact_agg_bookings["occ_pct"] = fact_agg_bookings["successful_bookings"]/fact_agg_bookings["capacity"]
fact_agg_bookings.head()
```

```python
fact_agg_bookings["occ_pct"] = fact_agg_bookings["occ_pct"].apply(lambda x:round(x*100, 2))
fact_agg_bookings.head()
```
